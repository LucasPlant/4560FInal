```python
import numpy as np
import so101
import math_utils
```

```python
# Desired end-effector transform: identity rotation, translation [0.1, 0.1, 0]
desired_wge = np.eye(4)
desired_wge[0:3, 3] = np.array([0.1, 0.1, 0.0])

# Call the inverse kinematics solver (returns joint angles in radians)
q_rad = so101.SO101._inverse_kinematics(desired_wge)
q_deg = np.rad2deg(q_rad)

print('Found joint angles (radians):', q_rad)
print('Found joint angles (degrees):', q_deg)

# Verify by forward kinematics using the internal rad-array FK
achieved_wge = so101.SO101._forward_kinematics(q_rad)

error = math_utils.tangent_space_error(achieved_wge, desired_wge)
error_norm = np.linalg.norm(error)
print('Tangent space error (twist):', error)
print('Tangent space error norm:', error_norm)
print('Achieved transform:\n', achieved_wge)
print('desired_wge:\n', desired_wge)
```
